```python
from scipy import stats
z_scores = np.abs(stats.zscore(df['serum_creatinine']))

# Flag values where z > 3
outliers_z = df[z_scores > 3]

print("Shape of outliers:", outliers_z.shape)
print(outliers_z[['serum_creatinine']])
```

```python
# For ejection_fraction
z_ef = np.abs(stats.zscore(df['ejection_fraction']))
outliers_ef = df[z_ef > 3]
print("Outliers in ejection_fraction:", outliers_ef.shape)

# For age
z_age = np.abs(stats.zscore(df['age']))
outliers_age = df[z_age > 3]
print("Outliers in age:", outliers_age.shape)
```

```python
#drop outliers
col='serum_creatinine'
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_iqr_removed = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("After IQR-based removal:", df_iqr_removed.shape)
```
